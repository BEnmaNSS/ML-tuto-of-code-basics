```python
import pandas as pd 
import numpy as np
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
#df is an array that shows the dimension 
digits.data[0:5]
```

```python
#see our array in datafram object 
df = pd.DataFrame(digits.data)
df
```

```python
digits.target
```

```python
df['target']= digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target', axis='columns'), df['target'], test_size=0.2)
print(X_train.shape) 
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
#here we can tune our hyperparameter to imporove accuracy 
model = RandomForestClassifier(n_estimators= 10)
#nestimator mean the number of trees
model = RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Now you can get one of the decision trees (e.g., the first tree)
chosen_tree = model.estimators_[0]

# Convert the unique class names from the target variable to strings
class_names = y_train.unique().astype(str).tolist()

# Increase the figure size
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=800)

# Increase the font size
plt.rcParams.update({'font.size': 10})

# Convert feature names to a list
feature_names = X_train.columns.tolist()

# Visualize the chosen tree
tree.plot_tree(chosen_tree,
               feature_names = feature_names, 
               class_names = class_names,
               filled = True)

# Save the figure
fig.savefig('random_forest.png')
```
